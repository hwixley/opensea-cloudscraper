```python
import cloudscraper as cs
import bs4
import json
import datetime
import pickle
import os
```

```python
scraper = cs.CloudScraper()  # CloudScraper inherits from requests.Session
url = "https://opensea.io/rankings"

soup = bs4.BeautifulSoup(scraper.get(url).text, "html.parser")
```

```python
offset = len("window.__wired__=")
script = soup.script.contents[0][offset:]

decoded_script = json.loads(script)["records"]
print(decoded_script)
```

```python
keys = decoded_script.keys()
count = 0
slugs = {}

for k in keys:
    if (k[-2:] == "==" or len(k) == 28) and "slug" in decoded_script[k].keys():
        count += 1
        slugs[decoded_script[k]["slug"]] = decoded_script[k]
        print(decoded_script[k]["slug"])

print(count)
```

```python
now = datetime.datetime.now()
dt_string = now.strftime("%d-%m-%Y_%H-%M-%S")

with open(f"{dt_string}_rankings.pkl", 'wb') as f:
    pickle.dump(slugs, f)
```
